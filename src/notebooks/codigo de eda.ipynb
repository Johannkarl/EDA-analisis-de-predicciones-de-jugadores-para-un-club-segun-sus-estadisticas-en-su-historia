```python
import pandas as pd

# Ruta correcta del archivo CSV
ruta = r'C:\Users\johan\OneDrive\Escritorio\EDA\dataset - 2020-09-24.csv'

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(ruta)

# Ver las primeras filas del DataFrame
df.head()
```

```python
# Verificar la estructura y tipos de datos del DataFrame
print(df.info())
```

```python
# Ver un resumen estadístico de las columnas numéricas
print(df.describe())
```

```python
# Ver las primeras filas del DataFrame
print(df.head())
```

```python
# Contar la cantidad de jugadores por club
jugadores_por_club = df['Club'].value_counts()

# Mostrar los 10 clubes con más jugadores
print(jugadores_por_club.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizar la distribución de jugadores por club (Top 10)
plt.figure(figsize=(12, 6))  # Ajustar el tamaño de la figura
sns.countplot(data=df, x='Club', order=jugadores_por_club.head(10).index, palette='viridis')

# Configuración del gráfico
plt.title('Distribución de Jugadores por Club (Top 10)', fontsize=16)
plt.xlabel('Club', fontsize=12)
plt.ylabel('Cantidad de Jugadores', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotar los nombres de los clubes para mejor visibilidad
plt.show()
```

```python
# Ranking de jugadores por goles
ranking_goles = df[['Name', 'Goals']].sort_values(by='Goals', ascending=False)

# Mostrar los 10 jugadores con más goles
print("Top 10 Jugadores con más goles:")
print(ranking_goles.head(10))

```

```python
#Distribucion de jugadores por posicion
import matplotlib.pyplot as plt
import seaborn as sns

# Contar jugadores por posición
jugadores_por_posicion = df['Position'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=jugadores_por_posicion.head(10).index, y=jugadores_por_posicion.head(10).values, palette='viridis')

# Configuración del gráfico
plt.title('Distribución de Jugadores por Posición (Top 10)', fontsize=16)
plt.xlabel('Posición', fontsize=12)
plt.ylabel('Cantidad de Jugadores', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotar los nombres de las posiciones para mejor visibilidad
plt.show()
```

Identificacion de talento

```python
# Filtrar jugadores con altos goles y asistencias
jugadores_destacados = df[['Name', 'Goals', 'Assists', 'Position']]

# Filtrar a los jugadores con al menos 10 goles y 10 asistencias
jugadores_destacados = jugadores_destacados[(jugadores_destacados['Goals'] >= 10) & (jugadores_destacados['Assists'] >= 10)]

# Mostrar los 10 mejores jugadores
print("Jugadores con más de 10 goles y 10 asistencias:")
print(jugadores_destacados[['Name', 'Goals', 'Assists']].sort_values(by=['Goals', 'Assists'], ascending=False).head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contar jugadores por posición
jugadores_por_posicion = df['Position'].value_counts()

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=jugadores_por_posicion.index, y=jugadores_por_posicion.values, palette='viridis')

# Título y etiquetas
plt.title('Distribución de Jugadores por Posición', fontsize=16)
plt.xlabel('Posición', fontsize=12)
plt.ylabel('Cantidad de Jugadores', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
# Esto solo es un complemento para ayudar en el analisis
```

```python
# Filtrar jugadores con más goles y asistencias
ranking_goles_asistencias = df[['Name', 'Goals', 'Assists', 'Position']]

# Filtrar a los jugadores con al menos 10 goles y 10 asistencias
jugadores_destacados = ranking_goles_asistencias[(ranking_goles_asistencias['Goals'] >= 10) & (ranking_goles_asistencias['Assists'] >= 10)]

# Mostrar los 10 jugadores con más goles y asistencias
print("Top 10 Jugadores con más de 10 goles y 10 asistencias:")
print(jugadores_destacados[['Name', 'Goals', 'Assists', 'Position']].sort_values(by=['Goals', 'Assists'], ascending=False).head(10))

# Crear gráfico de barras para los primeros 10 jugadores con más goles y asistencias
plt.figure(figsize=(12, 6))
sns.barplot(x=jugadores_destacados['Name'], y=jugadores_destacados['Goals'] + jugadores_destacados['Assists'], palette='coolwarm')

# Título y etiquetas
plt.title('Top 10 Jugadores con Más Goles y Asistencias', fontsize=16)
plt.xlabel('Jugadores', fontsize=12)
plt.ylabel('Total de Goles + Asistencias', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Filtrar jugadores con edad y goles/ asistencias
df_edad_rendimiento = df[['Age', 'Goals', 'Assists']].dropna()

# Crear gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Goals', data=df_edad_rendimiento, color='blue', label='Goles')
sns.scatterplot(x='Age', y='Assists', data=df_edad_rendimiento, color='red', label='Asistencias')

# Título y etiquetas
plt.title('Impacto de la Edad en los Goles y Asistencias', fontsize=16)
plt.xlabel('Edad', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
plt.legend()
plt.show()
```

```python
#
# Agrupar por Club y contar jugadores destacados (más de 10 goles y 10 asistencias)
jugadores_por_club = df[df['Goals'] >= 10 & (df['Assists'] >= 10)].groupby('Club')['Name'].count()

# Mostrar los clubes con más jugadores destacados
clubes_destacados = jugadores_por_club.sort_values(ascending=False).head(10)
print("Top 10 clubes con más jugadores destacados:")
print(clubes_destacados)

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=clubes_destacados.index, y=clubes_destacados.values, palette='magma')

# Título y etiquetas
plt.title('Top 10 Clubes con Más Jugadores Destacados', fontsize=16)
plt.xlabel('Club', fontsize=12)
plt.ylabel('Número de Jugadores Destacados', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
```
